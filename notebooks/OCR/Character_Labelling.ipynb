# Renaming Image Dataset

```python
import os

folder = "../../input/number_characters/9"   # your dataset directory
prefix = "9"                          # change if needed

files = sorted(os.listdir(folder))    # get files in order

counter = 1
for filename in files:
    old_path = os.path.join(folder, filename)

    # skip directories
    if not os.path.isfile(old_path):
        continue

    # new name with two digits (01, 02, 03)
    new_name = f"{prefix}-{counter:02d}.png"
    new_path = os.path.join(folder, new_name)

    os.rename(old_path, new_path)
    counter += 1

print("Renaming completed!")
```

# Create Dataset Metadata

```python
import os
import csv

root_dir = "../../input/number_characters"
output_csv = os.path.join(root_dir, "metadata.csv")

rows = []

# Loop over each label folder
for label in sorted(os.listdir(root_dir)):
    label_path = os.path.join(root_dir, label)

    # Skip non-directories
    if not os.path.isdir(label_path):
        continue

    # Loop inside each label folder
    for filename in sorted(os.listdir(label_path)):
        file_path = f"{label}/{filename}"     # relative path
        rows.append([file_path, label])        # append row: filepath, label

# Write CSV file
with open(output_csv, "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["filepath", "label"])    # header
    writer.writerows(rows)

print(f"metadata.csv created at {output_csv}")
```
