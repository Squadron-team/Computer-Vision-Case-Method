# Tesseract

```python
import pytesseract
from PIL import Image

text = pytesseract.image_to_string(Image.open("../../input/bni-receipts/fraud/f_025.png"))
print(text)
```

```python
import easyocr

reader = easyocr.Reader(['en', 'id'])
result = reader.readtext('screenshot.png', detail=0)
print(result)
```
